```python
# ============================================
# COMPLETE ENSEMBLE LEARNING PIPELINE
# Shrimp Disease Classification
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Input, concatenate
from tensorflow.keras.applications import VGG16, InceptionV3, ResNet50, EfficientNetB0, MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import plot_model
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from numpy import dstack, argmax
import time
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================
# CONFIGURATION - CHANGE ONLY THIS SECTION
# ============================================

# Just change the BASE_DIR to switch between datasets
#BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/Raw_Train_test_Validation"
BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/2_Conventional_augmented_2500_ttvs"

TRAIN_DIR = f"{BASE_DIR}/train"
VAL_DIR = f"{BASE_DIR}/validation"
TEST_DIR = f"{BASE_DIR}/test"

IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS_BASE = 30  # For base models (if training from scratch)
EPOCHS_META = 30  # For meta-learners
NUM_CLASSES = 4

print("=" * 80)
print("ENSEMBLE LEARNING PIPELINE - SHRIMP DISEASE CLASSIFICATION")
print("=" * 80)
print(f"\nDataset Directory: {BASE_DIR}")
print(f"Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"Batch Size: {BATCH_SIZE}")
print(f"Epochs for Meta-Models: {EPOCHS_META}")
print(f"Number of Classes: {NUM_CLASSES}")

# ============================================
# DATA PREPARATION
# ============================================
print("\n" + "=" * 80)
print("LOADING AND PREPARING DATA")
print("=" * 80)

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

validation_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print(f"\nTraining samples: {train_generator.samples}")
print(f"Validation samples: {validation_generator.samples}")
print(f"Test samples: {test_generator.samples}")
print(f"Classes: {list(train_generator.class_indices.keys())}")

# ============================================
# HELPER FUNCTIONS
# ============================================
def format_time(seconds):
    """Format seconds into readable time string"""
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    secs = int(seconds % 60)
    
    if hours > 0:
        return f"{hours}h {minutes}m {secs}s"
    elif minutes > 0:
        return f"{minutes}m {secs}s"
    else:
        return f"{secs}s"

def plot_training_history(history, model_name):
    """Plot training and validation accuracy and loss"""
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Accuracy plot
    axes[0].plot(history.history['accuracy'], linewidth=2, label='Train', marker='o')
    axes[0].plot(history.history['val_accuracy'], linewidth=2, label='Validation', marker='s')
    axes[0].set_title(f'{model_name} - Model Accuracy', fontsize=16, fontweight='bold')
    axes[0].set_xlabel('Epoch', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('Accuracy', fontsize=12, fontweight='bold')
    axes[0].legend(fontsize=12)
    axes[0].grid(True, alpha=0.3)

    # Loss plot
    axes[1].plot(history.history['loss'], linewidth=2, label='Train', marker='o')
    axes[1].plot(history.history['val_loss'], linewidth=2, label='Validation', marker='s')
    axes[1].set_title(f'{model_name} - Model Loss', fontsize=16, fontweight='bold')
    axes[1].set_xlabel('Epoch', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Loss', fontsize=12, fontweight='bold')
    axes[1].legend(fontsize=12)
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

def evaluate_model(model, test_generator, model_name):
    """Evaluate model on TEST data and display metrics"""
    print(f"\n{'=' * 80}")
    print(f"EVALUATION RESULTS ON TEST DATA - {model_name}")
    print("=" * 80)

    # Reset generator
    test_generator.reset()

    # Get predictions
    print("\n🔍 Generating predictions on test set...")
    predictions = model.predict(test_generator, verbose=1)
    y_pred = np.argmax(predictions, axis=1)
    y_true = test_generator.classes

    # Class names
    class_names = list(test_generator.class_indices.keys())

    # Calculate metrics
    accuracy = np.mean(y_pred == y_true)
    precision, recall, f1, support = precision_recall_fscore_support(
        y_true, y_pred, average=None, labels=range(NUM_CLASSES)
    )

    # Performance metrics table
    metrics_df = pd.DataFrame({
        'Class': class_names,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Support': support
    })

    # Overall metrics
    overall_precision, overall_recall, overall_f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='weighted'
    )

    overall_row = pd.DataFrame({
        'Class': ['Overall (Weighted Avg)'],
        'Precision': [overall_precision],
        'Recall': [overall_recall],
        'F1-Score': [overall_f1],
        'Support': [len(y_true)]
    })

    metrics_df = pd.concat([metrics_df, overall_row], ignore_index=True)

    print("\n📊 CLASSIFICATION PERFORMANCE METRICS (TEST SET):")
    print("=" * 80)
    print(metrics_df.to_string(index=False))
    print("=" * 80)
    print(f"\n🎯 Test Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

    # Classification Report
    print("\n" + "=" * 80)
    print("DETAILED CLASSIFICATION REPORT (TEST SET)")
    print("=" * 80)
    print("\n" + classification_report(y_true, y_pred, target_names=class_names, digits=4))

    # Confusion Matrix with enhanced formatting
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'}, 
                annot_kws={'size': 16, 'weight': 'bold'})
    plt.title(f'Confusion Matrix - {model_name} (Test Set)', fontsize=18, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=14, fontweight='bold')
    plt.xticks(fontsize=11, fontweight='bold', rotation=45, ha='right')
    plt.yticks(fontsize=11, fontweight='bold', rotation=0)
    plt.tight_layout()
    plt.show()

    return accuracy, overall_precision, overall_recall, overall_f1

def count_parameters(model):
    """Count total trainable and non-trainable parameters"""
    try:
        trainable = int(np.sum([tf.keras.backend.count_params(w) for w in model.trainable_weights]))
        non_trainable = int(np.sum([tf.keras.backend.count_params(w) for w in model.non_trainable_weights]))
        return trainable + non_trainable
    except:
        # Fallback for sklearn models
        return 0

# ============================================
# STORAGE FOR RESULTS
# ============================================
all_results = []

# ============================================
# CREATE BASE MODELS
# ============================================
def create_base_model(base_architecture, model_name, input_shape=(224, 224, 3)):
    """Create a base model with pretrained weights"""
    print(f"\n{'=' * 80}")
    print(f"BUILDING BASE MODEL: {model_name}")
    print("=" * 80)

    base = base_architecture(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )

    base.trainable = False

    inputs = Input(shape=input_shape)
    x = base(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(NUM_CLASSES, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs, name=model_name)

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print(f"\n✓ {model_name} built successfully!")
    return model

print("\n" + "=" * 80)
print("CREATING BASE MODELS")
print("=" * 80)

base_models_config = [
    (VGG16, 'VGG16'),
    (InceptionV3, 'InceptionV3'),
    (ResNet50, 'ResNet50'),
    (MobileNetV2, 'MobileNetV2'),
    (EfficientNetB0, 'EfficientNetB0')
]

base_models = []
base_model_names = []

for architecture, name in base_models_config:
    model = create_base_model(architecture, name)
    base_models.append(model)
    base_model_names.append(name)

print(f"\n✓ Created {len(base_models)} base models")

# ============================================
# GENERATE BASE MODEL PREDICTIONS
# ============================================
print("\n" + "=" * 80)
print("GENERATING BASE MODEL PREDICTIONS")
print("=" * 80)

# Get validation data for meta-learning
print("\nLoading validation data...")
validation_generator.reset()
val_images = []
val_labels = []

for i in range(len(validation_generator)):
    x_batch, y_batch = validation_generator[i]
    val_images.append(x_batch)
    val_labels.append(y_batch)

val_images = np.vstack(val_images)
val_labels = np.vstack(val_labels)
val_labels_int = np.argmax(val_labels, axis=1)

print(f"Validation samples loaded: {len(val_labels)}")

# Split validation into holdout (for meta-learner training) and meta-test
holdout_images, metatest_images, holdout_labels, metatest_labels = train_test_split(
    val_images, val_labels_int, test_size=0.5, random_state=42, stratify=val_labels_int
)

print(f"Holdout set: {len(holdout_labels)} samples (for training meta-learner)")
print(f"Meta-test set: {len(metatest_labels)} samples (for meta-learner validation)")

# Get test data
print("\nLoading test data...")
test_generator.reset()
test_images = []
test_labels = []

for i in range(len(test_generator)):
    x_batch, y_batch = test_generator[i]
    test_images.append(x_batch)
    test_labels.append(y_batch)

test_images = np.vstack(test_images)
test_labels = np.vstack(test_labels)
test_labels_int = np.argmax(test_labels, axis=1)

print(f"Test samples loaded: {len(test_labels)}")

# Generate predictions from base models
print("\n--- Generating Predictions from Base Models ---")
holdout_preds_list = []
metatest_preds_list = []
test_preds_list = []

for i, model in enumerate(base_models):
    print(f"\n{base_model_names[i]}:")
    
    print("  - Holdout predictions...")
    holdout_pred = model.predict(holdout_images, verbose=0)
    holdout_preds_list.append(holdout_pred)
    
    print("  - Meta-test predictions...")
    metatest_pred = model.predict(metatest_images, verbose=0)
    metatest_preds_list.append(metatest_pred)
    
    print("  - Test predictions...")
    test_pred = model.predict(test_images, verbose=0)
    test_preds_list.append(test_pred)

# Stack predictions
holdout_preds_stacked = np.hstack(holdout_preds_list)  # Shape: (n_samples, 5*4)
metatest_preds_stacked = np.hstack(metatest_preds_list)
test_preds_stacked = np.hstack(test_preds_list)

print(f"\nHoldout predictions shape: {holdout_preds_stacked.shape}")
print(f"Meta-test predictions shape: {metatest_preds_stacked.shape}")
print(f"Test predictions shape: {test_preds_stacked.shape}")

# ============================================
# ENSEMBLE 1: SIMPLE AVERAGING
# ============================================
print("\n" + "#" * 80)
print("ENSEMBLE 1: SIMPLE AVERAGING")
print("#" * 80)

start_time = time.time()

# Average predictions across all base models
avg_holdout = np.mean(np.array(holdout_preds_list), axis=0)
avg_test = np.mean(np.array(test_preds_list), axis=0)

# Predictions
y_pred_avg = np.argmax(avg_test, axis=1)
avg_accuracy = accuracy_score(test_labels_int, y_pred_avg)

end_time = time.time()
total_time = end_time - start_time

print(f"\n✓ Simple Averaging Accuracy on Test Set: {avg_accuracy:.4f}")

# Evaluate
class_names = list(test_generator.class_indices.keys())
precision, recall, f1, _ = precision_recall_fscore_support(
    test_labels_int, y_pred_avg, average='weighted'
)

# Confusion Matrix
cm = confusion_matrix(test_labels_int, y_pred_avg)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names,
            cbar_kws={'label': 'Count'}, 
            annot_kws={'size': 16, 'weight': 'bold'})
plt.title('Confusion Matrix - Simple Averaging Ensemble (Test Set)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontsize=14, fontweight='bold')
plt.xticks(fontsize=11, fontweight='bold', rotation=45, ha='right')
plt.yticks(fontsize=11, fontweight='bold', rotation=0)
plt.tight_layout()
plt.show()

all_results.append({
    'Ensemble Type': 'Averaging',
    'Meta-Model': 'Simple Average',
    'Accuracy': avg_accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1,
    'Total Parameters': 0,
    'Training Time': format_time(total_time)
})

# ============================================
# ARCHITECTURAL DIAGRAM - AVERAGING
# ============================================
print("\n" + "=" * 80)
print("GENERATING ARCHITECTURAL DIAGRAM - AVERAGING ENSEMBLE")
print("=" * 80)

# Create a simple functional model to represent averaging
avg_inputs = [Input(shape=(NUM_CLASSES,), name=f'{name}_input') for name in base_model_names]
avg_layer = tf.keras.layers.Average(name='averaging_layer')(avg_inputs)
avg_viz_model = Model(inputs=avg_inputs, outputs=avg_layer, name='Averaging_Ensemble')

plot_model(avg_viz_model, to_file='averaging_ensemble_architecture.png', 
           show_shapes=True, show_layer_names=True, rankdir='TB', dpi=150)
print("✓ Averaging ensemble architecture saved to 'averaging_ensemble_architecture.png'")

# ============================================
# ENSEMBLE 2: BLENDING WITH MULTIPLE META-MODELS
# ============================================
print("\n" + "#" * 80)
print("ENSEMBLE 2: BLENDING")
print("#" * 80)

blending_meta_models = [
    ('Logistic Regression', 'sklearn'),
    ('Dense_64_32', [64, 32]),
    ('Dense_128_64', [128, 64]),
    ('Dense_256_128_64', [256, 128, 64])
]

for meta_name, meta_config in blending_meta_models:
    print(f"\n{'*' * 80}")
    print(f"BLENDING - Meta-Model: {meta_name}")
    print('*' * 80)
    
    start_time = time.time()
    
    if meta_config == 'sklearn':
        # Sklearn model
        meta_model = LogisticRegression(random_state=42, max_iter=1000)
        meta_model.fit(holdout_preds_stacked, holdout_labels)
        
        # Predict
        test_pred_proba = meta_model.predict_proba(test_preds_stacked)
        y_pred_blend = np.argmax(test_pred_proba, axis=1)
        
        total_params = 0
        
    else:
        # Neural network meta-model
        layer_sizes = meta_config
        
        meta_input = Input(shape=(holdout_preds_stacked.shape[1],))
        x = meta_input
        
        for i, size in enumerate(layer_sizes):
            x = Dense(size, activation='relu')(x)
            if i < len(layer_sizes) - 1:
                x = Dropout(0.3)(x)
        
        meta_output = Dense(NUM_CLASSES, activation='softmax')(x)
        meta_model = Model(inputs=meta_input, outputs=meta_output)
        
        meta_model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print(f"\nMeta-model architecture ({meta_name}):")
        meta_model.summary()
        
        # Train
        from tensorflow.keras.utils import to_categorical
        holdout_labels_cat = to_categorical(holdout_labels, NUM_CLASSES)
        metatest_labels_cat = to_categorical(metatest_labels, NUM_CLASSES)
        
        history = meta_model.fit(
            holdout_preds_stacked,
            holdout_labels_cat,
            validation_data=(metatest_preds_stacked, metatest_labels_cat),
            epochs=EPOCHS_META,
            batch_size=32,
            verbose=1
        )
        
        # Plot training history
        plot_training_history(history, f"Blending - {meta_name}")
        
        # Predict
        test_pred_proba = meta_model.predict(test_preds_stacked, verbose=0)
        y_pred_blend = np.argmax(test_pred_proba, axis=1)
        
        total_params = count_parameters(meta_model)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    # Evaluate
    blend_accuracy = accuracy_score(test_labels_int, y_pred_blend)
    precision, recall, f1, _ = precision_recall_fscore_support(
        test_labels_int, y_pred_blend, average='weighted'
    )
    
    print(f"\n✓ {meta_name} Blending Accuracy: {blend_accuracy:.4f}")
    
    # Confusion Matrix
    cm = confusion_matrix(test_labels_int, y_pred_blend)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'}, 
                annot_kws={'size': 16, 'weight': 'bold'})
    plt.title(f'Confusion Matrix - Blending ({meta_name}) - Test Set', 
              fontsize=18, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=14, fontweight='bold')
    plt.xticks(fontsize=11, fontweight='bold', rotation=45, ha='right')
    plt.yticks(fontsize=11, fontweight='bold', rotation=0)
    plt.tight_layout()
    plt.show()
    
    all_results.append({
        'Ensemble Type': 'Blending',
        'Meta-Model': meta_name,
        'Accuracy': blend_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Total Parameters': total_params,
        'Training Time': format_time(total_time)
    })

# ============================================
# ARCHITECTURAL DIAGRAM - BLENDING
# ============================================
print("\n" + "=" * 80)
print("GENERATING ARCHITECTURAL DIAGRAM - BLENDING ENSEMBLE")
print("=" * 80)

# Create blending architecture visualization
blend_inputs = [Input(shape=(NUM_CLASSES,), name=f'{name}_output') for name in base_model_names]
blend_concat = concatenate(blend_inputs, name='concatenate_predictions')
blend_hidden = Dense(64, activation='relu', name='meta_hidden')(blend_concat)
blend_output = Dense(NUM_CLASSES, activation='softmax', name='meta_output')(blend_hidden)
blend_viz_model = Model(inputs=blend_inputs, outputs=blend_output, name='Blending_Ensemble')

plot_model(blend_viz_model, to_file='blending_ensemble_architecture.png', 
           show_shapes=True, show_layer_names=True, rankdir='TB', dpi=150)
print("✓ Blending ensemble architecture saved to 'blending_ensemble_architecture.png'")

# ============================================
# ENSEMBLE 3: STACKING WITH MULTIPLE META-MODELS
# ============================================
print("\n" + "#" * 80)
print("ENSEMBLE 3: STACKING (INTEGRATED)")
print("#" * 80)

stacking_meta_models = [
    ('Dense_64_32', [64, 32]),
    ('Dense_128_64_32', [128, 64, 32]),
    ('Dense_256_128', [256, 128]),
    ('Dense_512_256_128', [512, 256, 128])
]

def create_stacked_model(base_models, layer_sizes, meta_name):
    """Create integrated stacking model"""
    
    # Create new input
    stacked_input = Input(shape=(IMG_SIZE, IMG_SIZE, 3), name='stacked_input')
    
    # Get outputs from all base models (frozen)
    ensemble_outputs = []
    for i, base_model in enumerate(base_models):
        # Make layers non-trainable
        for layer in base_model.layers:
            layer.trainable = False
        
        # Get output from this model
        x = base_model(stacked_input)
        ensemble_outputs.append(x)
    
    # Concatenate all outputs
    merged = concatenate(ensemble_outputs, name='concat_base_outputs')
    
    # Meta-learner layers
    x = merged
    for i, size in enumerate(layer_sizes):
        x = Dense(size, activation='relu', name=f'meta_dense_{i+1}')(x)
        if i < len(layer_sizes) - 1:
            x = Dropout(0.3, name=f'meta_dropout_{i+1}')(x)
    
    # Final output
    output = Dense(NUM_CLASSES, activation='softmax', name='final_output')(x)
    
    # Create model
    model = Model(inputs=stacked_input, outputs=output, name=f'Stacked_{meta_name}')
    
    return model

for meta_name, layer_sizes in stacking_meta_models:
    print(f"\n{'*' * 80}")
    print(f"STACKING - Meta-Model: {meta_name}")
    print('*' * 80)
    
    start_time = time.time()
    
    # Create stacked model
    stacked_model = create_stacked_model(base_models, layer_sizes, meta_name)
    
    stacked_model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print(f"\nStacked model architecture ({meta_name}):")
    stacked_model.summary()
    
    # Prepare data
    from tensorflow.keras.utils import to_categorical
    holdout_labels_cat = to_categorical(holdout_labels, NUM_CLASSES)
    metatest_labels_cat = to_categorical(metatest_labels, NUM_CLASSES)
    
    # Train
    history = stacked_model.fit(
        holdout_images,
        holdout_labels_cat,
        validation_data=(metatest_images, metatest_labels_cat),
        epochs=EPOCHS_META,
        batch_size=BATCH_SIZE,
        verbose=1
    )
    
    # Plot training history
    plot_training_history(history, f"Stacking - {meta_name}")
    
    # Predict on test set
    test_pred_proba = stacked_model.predict(test_images, verbose=0)
    y_pred_stack = np.argmax(test_pred_proba, axis=1)
    
    end_time = time.time()
    total_time = end_time - start_time
    
    # Evaluate
    stack_accuracy = accuracy_score(test_labels_int, y_pred_stack)
    precision, recall, f1, _ = precision_recall_fscore_support(
        test_labels_int, y_pred_stack, average='weighted'
    )
    
    total_params = count_parameters(stacked_model)
    
    print(f"\n✓ {meta_name} Stacking Accuracy: {stack_accuracy:.4f}")
    
    # Confusion Matrix
    cm = confusion_matrix(test_labels_int, y_pred_stack)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'}, 
                annot_kws={'size': 16, 'weight': 'bold'})
    plt.title(f'Confusion Matrix - Stacking ({meta_name}) - Test Set', 
              fontsize=18, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=14, fontweight='bold')
    plt.xticks(fontsize=11, fontweight='bold', rotation=45, ha='right')
    plt.yticks(fontsize=11, fontweight='bold', rotation=0)
    plt.tight_layout()
    plt.show()
    
    all_results.append({
        'Ensemble Type': 'Stacking',
        'Meta-Model': meta_name,
        'Accuracy': stack_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Total Parameters': total_params,
        'Training Time': format_time(total_time)
    })

# ============================================
# ARCHITECTURAL DIAGRAM - STACKING
# ============================================
print("\n" + "=" * 80)
print("GENERATING ARCHITECTURAL DIAGRAM - STACKING ENSEMBLE")
print("=" * 80)

# Create a representative stacking model for visualization
stack_viz_input = Input(shape=(IMG_SIZE, IMG_SIZE, 3), name='input_image')

# Simplified base model representations
base_outputs_viz = []
for name in base_model_names:
    x = Dense(NUM_CLASSES, activation='softmax', name=f'{name}_output')(stack_viz_input)
    base_outputs_viz.append(x)

stack_concat = concatenate(base_outputs_viz, name='concat_all_predictions')
stack_meta_hidden = Dense(128, activation='relu', name='meta_learner_hidden')(stack_concat)
stack_meta_output = Dense(NUM_CLASSES, activation='softmax', name='final_prediction')(stack_meta_hidden)

stack_viz_model = Model(inputs=stack_viz_input, outputs=stack_meta_output, name='Stacking_Ensemble')

plot_model(stack_viz_model, to_file='stacking_ensemble_architecture.png', 
           show_shapes=True, show_layer_names=True, rankdir='TB', dpi=150)
print("✓ Stacking ensemble architecture saved to 'stacking_ensemble_architecture.png'")

# ============================================
# FINAL RESULTS COMPARISON
# ============================================
print("\n" + "=" * 80)
print("FINAL RESULTS COMPARISON")
print("=" * 80)

results_df = pd.DataFrame(all_results)

print("\n📊 ENSEMBLE METHODS PERFORMANCE COMPARISON:")
print("=" * 80)
print(results_df.to_string(index=False))
print("=" * 80)

# Save to CSV
csv_filename = 'ensemble_results_comparison.csv'
results_df.to_csv(csv_filename, index=False)
print(f"\n✓ Results saved to '{csv_filename}'")

# Find best model
best_idx = results_df['Accuracy'].idxmax()
best_model = results_df.iloc[best_idx]

print(f"\n🏆 BEST ENSEMBLE MODEL:")
print(f"   Type: {best_model['Ensemble Type']}")
print(f"   Meta-Model: {best_model['Meta-Model']}")
print(f"   Accuracy: {best_model['Accuracy']:.4f} ({best_model['Accuracy']*100:.2f}%)")
print(f"   Precision: {best_model['Precision']:.4f}")
print(f"   Recall: {best_model['Recall']:.4f}")
print(f"   F1-Score: {best_model['F1-Score']:.4f}")
print(f"   Training Time: {best_model['Training Time']}")

# Visualization of results
fig, axes = plt.subplots(2, 2, figsize=(20, 14))

# Accuracy comparison
ax1 = axes[0, 0]
bars1 = ax1.bar(range(len(results_df)), results_df['Accuracy'], 
                color=['steelblue', 'lightcoral', 'lightcoral', 'lightcoral', 'lightcoral',
                       'lightgreen', 'lightgreen', 'lightgreen', 'lightgreen'])
ax1.set_xticks(range(len(results_df)))
ax1.set_xticklabels([f"{row['Ensemble Type']}\n{row['Meta-Model']}" 
                      for _, row in results_df.iterrows()], 
                     rotation=45, ha='right', fontsize=9)
ax1.set_title('Accuracy Comparison Across Ensembles', fontsize=14, fontweight='bold')
ax1.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)
for i, bar in enumerate(bars1):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.4f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

# F1-Score comparison
ax2 = axes[0, 1]
bars2 = ax2.bar(range(len(results_df)), results_df['F1-Score'],
                color=['steelblue', 'lightcoral', 'lightcoral', 'lightcoral', 'lightcoral',
                       'lightgreen', 'lightgreen', 'lightgreen', 'lightgreen'])
ax2.set_xticks(range(len(results_df)))
ax2.set_xticklabels([f"{row['Ensemble Type']}\n{row['Meta-Model']}" 
                      for _, row in results_df.iterrows()], 
                     rotation=45, ha='right', fontsize=9)
ax2.set_title('F1-Score Comparison Across Ensembles', fontsize=14, fontweight='bold')
ax2.set_ylabel('F1-Score', fontsize=12, fontweight='bold')
ax2.grid(axis='y', alpha=0.3)
for i, bar in enumerate(bars2):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.4f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

# Precision vs Recall
ax3 = axes[1, 0]
scatter_colors = ['blue' if t == 'Averaging' else 'red' if t == 'Blending' else 'green' 
                  for t in results_df['Ensemble Type']]
ax3.scatter(results_df['Precision'], results_df['Recall'], 
           c=scatter_colors, s=200, alpha=0.6, edgecolors='black', linewidth=2)
for i, row in results_df.iterrows():
    ax3.annotate(f"{row['Meta-Model'][:10]}", 
                (row['Precision'], row['Recall']),
                fontsize=8, ha='right')
ax3.set_xlabel('Precision', fontsize=12, fontweight='bold')
ax3.set_ylabel('Recall', fontsize=12, fontweight='bold')
ax3.set_title('Precision vs Recall', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.legend(['Averaging', 'Blending', 'Stacking'], fontsize=10)

# Parameters comparison (excluding averaging)
ax4 = axes[1, 1]
params_df = results_df[results_df['Total Parameters'] > 0]
bars4 = ax4.bar(range(len(params_df)), params_df['Total Parameters'],
                color=['lightcoral']*3 + ['lightgreen']*4)
ax4.set_xticks(range(len(params_df)))
ax4.set_xticklabels([f"{row['Ensemble Type']}\n{row['Meta-Model']}" 
                      for _, row in params_df.iterrows()], 
                     rotation=45, ha='right', fontsize=9)
ax4.set_title('Model Parameters Comparison', fontsize=14, fontweight='bold')
ax4.set_ylabel('Total Parameters', fontsize=12, fontweight='bold')
ax4.grid(axis='y', alpha=0.3)
for i, bar in enumerate(bars4):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height):,}', ha='center', va='bottom', fontsize=7, fontweight='bold', rotation=45)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("ENSEMBLE LEARNING PIPELINE COMPLETED!")
print("=" * 80)
print(f"\nGenerated Files:")
print(f"  1. averaging_ensemble_architecture.png")
print(f"  2. blending_ensemble_architecture.png")
print(f"  3. stacking_ensemble_architecture.png")
print(f"  4. {csv_filename}")
```

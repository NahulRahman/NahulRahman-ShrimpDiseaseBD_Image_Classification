```python
#!/usr/bin/env python3
"""
===================================================================================
PRODUCTION-READY XAI PIPELINE FOR SHRIMP DISEASE CLASSIFICATION
===================================================================================
Explainability Analysis for "Healthy" Class
- 6 XAI Methods: GradCAM, GradCAM++, LIME, Integrated Gradients, Occlusion, Saliency
- Comparative analysis across RAW, Augmented, and GAN-augmented images
- Single-instance explainability with quantitative metrics
- MobileNetV2 with 250x250 input, 30 epochs, batch size 64
===================================================================================
"""

# ===================================================================================
# PACKAGE INSTALLATION (RUN THIS FIRST IN KAGGLE)
# ===================================================================================
"""
KAGGLE INSTALLATION COMMANDS:
Copy and paste this in a Kaggle notebook cell:

!pip install scikit-image opencv-python tqdm --quiet
"""

# Suppress warnings
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import cv2
from scipy.ndimage import gaussian_filter
import random
from skimage.segmentation import mark_boundaries, slic
from tqdm import tqdm
import time

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

print("=" * 100)
print("XAI PIPELINE - SHRIMP DISEASE CLASSIFICATION (HEALTHY CLASS)")
print("=" * 100)

# ===================================================================================
# CONFIGURATION - MOBILENETV2 PARAMETERS
# ===================================================================================

# MobileNetV2 Fixed Parameters (DO NOT CHANGE)
BATCH_SIZE = 64
EPOCHS = 30
IMG_SIZE = (250, 250)  # Fixed 250x250 for all models
NUM_CLASSES = 4

# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# DATASET SELECTION - UNCOMMENT THE ONE YOU WANT TO USE
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

# OPTION 1: RAW DATASET
DATASET_TYPE = "RAW"
BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/Raw_Train_test_Validation"
AUGMENTATION_MARKER = None  # No marker for raw images

# OPTION 2: CONVENTIONAL AUGMENTED DATASET (Uncomment below, comment above)
# DATASET_TYPE = "CONVENTIONAL_AUGMENTED"
# BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/2_Conventional_augmented_2500_ttvs"
# AUGMENTATION_MARKER = "-conv-aug-"  # Marker to identify augmented images

# OPTION 3: GAN AUGMENTED DATASET (Add your path)
# DATASET_TYPE = "GAN_AUGMENTED"
# BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/GAN_augmented_dataset"
# AUGMENTATION_MARKER = "-gan-aug-"  # Marker to identify GAN augmented images

# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

# Paths
TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR = os.path.join(BASE_DIR, "validation")
TEST_DIR = os.path.join(BASE_DIR, "test")

TRAIN_HEALTHY = os.path.join(TRAIN_DIR, "Healthy")
VAL_HEALTHY = os.path.join(VAL_DIR, "Healthy")
TEST_HEALTHY = os.path.join(TEST_DIR, "Healthy")

# Output paths
OUTPUT_DIR = f"/home/claude/xai_outputs_{DATASET_TYPE.lower()}"
MODELS_DIR = os.path.join(OUTPUT_DIR, "models")
XAI_DIR = os.path.join(OUTPUT_DIR, "xai_results")
METRICS_DIR = os.path.join(OUTPUT_DIR, "metrics")

for directory in [OUTPUT_DIR, MODELS_DIR, XAI_DIR, METRICS_DIR]:
    os.makedirs(directory, exist_ok=True)

print(f"\nüìã CONFIGURATION:")
print(f"   Dataset Type: {DATASET_TYPE}")
print(f"   Base Directory: {BASE_DIR}")
print(f"   Image Size: {IMG_SIZE}")
print(f"   Batch Size: {BATCH_SIZE}")
print(f"   Epochs: {EPOCHS}")
print(f"   Number of Classes: {NUM_CLASSES}")
print(f"\nüìÅ OUTPUT:")
print(f"   Output Directory: {OUTPUT_DIR}")
print(f"   Models: {MODELS_DIR}")
print(f"   XAI Results: {XAI_DIR}")
print(f"   Metrics: {METRICS_DIR}")

# ===================================================================================
# UTILITY FUNCTIONS
# ===================================================================================

def get_healthy_images(directory, augmentation_marker=None, n_samples=None):
    """Get list of healthy class images"""
    if not os.path.exists(directory):
        print(f"‚ö† Warning: Directory not found: {directory}")
        return []
    
    all_files = [f for f in os.listdir(directory) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    if augmentation_marker:
        filtered_files = [f for f in all_files if augmentation_marker in f]
    else:
        filtered_files = [f for f in all_files if '-conv-aug-' not in f and '-gan-aug-' not in f]
    
    if n_samples:
        filtered_files = random.sample(filtered_files, min(n_samples, len(filtered_files)))
    
    return [os.path.join(directory, f) for f in filtered_files]

def load_and_preprocess_image(img_path, target_size=IMG_SIZE):
    """Load and preprocess image for model"""
    img = load_img(img_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    return img_array, img

def display_image(img_array, title="", figsize=(6, 6)):
    """Display image"""
    plt.figure(figsize=figsize)
    if len(img_array.shape) == 4:
        img_array = img_array[0]
    if img_array.shape[-1] == 3:
        plt.imshow(img_array)
    else:
        plt.imshow(img_array, cmap='gray')
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# ===================================================================================
# LOAD/TRAIN MOBILENETV2 MODEL
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 1: LOADING/TRAINING MOBILENETV2 MODEL")
print("=" * 100)

model_path = os.path.join(MODELS_DIR, 'mobilenetv2_shrimp_250x250.h5')

if os.path.exists(model_path):
    print(f"\nLoading existing model from {model_path}")
    model = load_model(model_path)
    print("‚úì Model loaded successfully")
else:
    print("\nModel not found. Training new MobileNetV2 model...")
    print(f"Parameters: IMG_SIZE={IMG_SIZE}, BATCH_SIZE={BATCH_SIZE}, EPOCHS={EPOCHS}")
    
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    
    val_datagen = ImageDataGenerator(rescale=1./255)
    
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True
    )
    
    val_generator = val_datagen.flow_from_directory(
        VAL_DIR,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )
    
    print(f"\n‚úì Data generators created")
    print(f"  Training samples: {train_generator.samples}")
    print(f"  Validation samples: {val_generator.samples}")
    
    base_model = MobileNetV2(
        input_shape=IMG_SIZE + (3,),
        include_top=False,
        weights='imagenet'
    )
    
    for layer in base_model.layers:
        layer.trainable = False
    
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    predictions = Dense(NUM_CLASSES, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    
    model.compile(
        optimizer=Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    total_params = model.count_params()
    print(f"\n‚úì MobileNetV2 model compiled")
    print(f"  Total parameters: {total_params:,}")
    print(f"  Input shape: {IMG_SIZE + (3,)}")
    
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
    ]
    
    print("\nüöÄ Training MobileNetV2...")
    training_start = time.time()
    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=EPOCHS,
        callbacks=callbacks,
        verbose=1
    )
    training_time = time.time() - training_start
    
    print(f"\n‚úì Training complete in {training_time/60:.2f} minutes")
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].plot(history.history['accuracy'], label='Train', linewidth=2, marker='o')
    axes[0].plot(history.history['val_accuracy'], label='Validation', linewidth=2, marker='s')
    axes[0].set_title('Model Accuracy', fontsize=13, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    axes[1].plot(history.history['loss'], label='Train', linewidth=2, marker='o')
    axes[1].plot(history.history['val_loss'], label='Validation', linewidth=2, marker='s')
    axes[1].set_title('Model Loss', fontsize=13, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.suptitle(f'MobileNetV2 Training History - {DATASET_TYPE}', fontsize=15, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    model.save(model_path)
    print(f"\n‚úì Model saved to {model_path}")

# Get class indices
try:
    temp_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
        TRAIN_DIR, target_size=IMG_SIZE, batch_size=1, class_mode='categorical'
    )
    class_indices = temp_gen.class_indices
except:
    class_indices = {'Healthy': 0, 'IHHNV': 1, 'WSSV': 2, 'YHV': 3}

idx_to_class = {v: k for k, v in class_indices.items()}
HEALTHY_CLASS_IDX = class_indices['Healthy']

print(f"\nüìä Class Information:")
print(f"   Class indices: {class_indices}")
print(f"   Healthy class index: {HEALTHY_CLASS_IDX}")

# ===================================================================================
# SELECT SAMPLE IMAGES
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 2: SELECTING SAMPLE IMAGES")
print("=" * 100)

train_samples = get_healthy_images(TRAIN_HEALTHY, AUGMENTATION_MARKER, n_samples=1)
val_samples = get_healthy_images(VAL_HEALTHY, AUGMENTATION_MARKER, n_samples=1)
test_samples = get_healthy_images(TEST_HEALTHY, AUGMENTATION_MARKER, n_samples=1)

initial_samples = train_samples + val_samples + test_samples
print(f"\nSelected {len(initial_samples)} initial samples:")
for sample in initial_samples:
    print(f"  - {os.path.basename(sample)}")

all_healthy = get_healthy_images(VAL_HEALTHY, AUGMENTATION_MARKER)
if len(all_healthy) == 0:
    all_healthy = get_healthy_images(TEST_HEALTHY, AUGMENTATION_MARKER)
if len(all_healthy) == 0:
    all_healthy = get_healthy_images(TRAIN_HEALTHY, AUGMENTATION_MARKER)

if len(all_healthy) > 0:
    ANALYSIS_IMAGE = random.choice(all_healthy)
    print(f"\nüì∏ Selected image for detailed XAI analysis:")
    print(f"   {os.path.basename(ANALYSIS_IMAGE)}")
else:
    print("‚ö† Warning: No healthy images found!")
    ANALYSIS_IMAGE = None

# ===================================================================================
# INITIAL VISUALIZATION
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 3: INITIAL VISUALIZATION - SAMPLES FROM EACH SPLIT")
print("=" * 100)

if len(initial_samples) >= 3:
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    splits = ['Train', 'Validation', 'Test']
    
    for idx, (img_path, split_name) in enumerate(zip(initial_samples, splits)):
        img, _ = load_and_preprocess_image(img_path)
        pred = model.predict(img, verbose=0)
        pred_class = idx_to_class[np.argmax(pred[0])]
        confidence = np.max(pred[0]) * 100
        
        axes[idx].imshow(img[0])
        axes[idx].set_title(f'{split_name}\n{os.path.basename(img_path)}\nPred: {pred_class} ({confidence:.1f}%)',
                           fontsize=11, fontweight='bold')
        axes[idx].axis('off')
    
    plt.suptitle(f'Healthy Class Samples - {DATASET_TYPE}', fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

if ANALYSIS_IMAGE is None:
    print("\n‚ö† ERROR: No analysis image available. Exiting.")
    exit()

# ===================================================================================
# XAI METHOD 1: GRADCAM
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 1: GRAD-CAM")
print("=" * 100)

def get_last_conv_layer(model):
    """Find last convolutional layer"""
    for layer in reversed(model.layers):
        if 'conv' in layer.name.lower():
            return layer.name
    return None

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """Generate GradCAM heatmap"""
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_index]
    
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def apply_gradcam(img_array, heatmap, original_img):
    """Apply GradCAM heatmap to image"""
    if isinstance(original_img, np.ndarray):
        if original_img.max() <= 1.0:
            original_img = original_img * 255
        original_img = original_img.astype('uint8')
    
    target_h, target_w = original_img.shape[:2]
    heatmap_resized = cv2.resize(heatmap, (target_w, target_h))
    heatmap_resized = np.uint8(255 * heatmap_resized)
    heatmap_colored = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)
    
    superimposed = heatmap_colored * 0.4 + original_img
    superimposed = np.clip(superimposed, 0, 255).astype('uint8')
    return superimposed

last_conv = get_last_conv_layer(model)
print(f"Last convolutional layer: {last_conv}")

img_array, original_img = load_and_preprocess_image(ANALYSIS_IMAGE)
original_img_np = np.array(original_img)

pred = model.predict(img_array, verbose=0)
pred_class_idx = np.argmax(pred[0])
confidence = pred[0][pred_class_idx]

print(f"\nPrediction: {idx_to_class[pred_class_idx]} (Confidence: {confidence:.4f})")

heatmap_healthy = make_gradcam_heatmap(img_array, model, last_conv, HEALTHY_CLASS_IDX)
gradcam_img = apply_gradcam(img_array[0], heatmap_healthy, original_img_np)

print("\nüìä GradCAM Results:")
display_image(img_array, "Original Image")
display_image(gradcam_img / 255.0, f"GradCAM - Healthy Class\nConfidence: {confidence:.4f}")

# ===================================================================================
# XAI METHOD 2: GRAD-CAM++
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 2: GRAD-CAM++")
print("=" * 100)

def make_gradcam_plusplus_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """Generate GradCAM++ heatmap"""
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    
    with tf.GradientTape() as tape1:
        with tf.GradientTape() as tape2:
            with tf.GradientTape() as tape3:
                conv_outputs, predictions = grad_model(img_array)
                if pred_index is None:
                    pred_index = tf.argmax(predictions[0])
                class_channel = predictions[:, pred_index]
            grads = tape3.gradient(class_channel, conv_outputs)
        grads_2 = tape2.gradient(grads, conv_outputs)
    grads_3 = tape1.gradient(grads_2, conv_outputs)
    
    grads = grads[0]
    grads_2 = grads_2[0]
    grads_3 = grads_3[0]
    
    global_sum = tf.reduce_sum(conv_outputs[0], axis=(0, 1), keepdims=True)
    alpha_denom = grads_2 * 2.0 + grads_3 * global_sum
    alpha_denom = tf.where(alpha_denom != 0.0, alpha_denom, tf.ones_like(alpha_denom))
    alphas = grads_2 / alpha_denom
    
    weights = tf.reduce_sum(alphas * tf.nn.relu(grads), axis=(0, 1))
    
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ weights[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.nn.relu(heatmap)
    heatmap = heatmap / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

heatmap_healthy_pp = make_gradcam_plusplus_heatmap(img_array, model, last_conv, HEALTHY_CLASS_IDX)
gradcam_pp_img = apply_gradcam(img_array[0], heatmap_healthy_pp, original_img_np)

print("\nüìä GradCAM++ Results:")
display_image(gradcam_pp_img / 255.0, f"GradCAM++ - Healthy Class\nConfidence: {confidence:.4f}")

# ===================================================================================
# XAI METHOD 3: INTEGRATED GRADIENTS - FIXED
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 3: INTEGRATED GRADIENTS")
print("=" * 100)

def integrated_gradients(img_array, model, target_class_idx, baseline=None, num_steps=50):
    """Compute integrated gradients - FIXED VERSION"""
    if baseline is None:
        baseline = np.zeros_like(img_array)
    
    alphas = np.linspace(0, 1, num_steps)
    gradients = []
    
    print(f"Computing {num_steps} interpolated gradients...")
    for alpha in tqdm(alphas, desc="IG Progress"):
        # Interpolate between baseline and image
        interpolated = baseline + alpha * (img_array - baseline)
        
        # FIXED: Don't add extra dimension - interpolated is already (1, H, W, C)
        interpolated_tensor = tf.convert_to_tensor(interpolated, dtype=tf.float32)
        
        with tf.GradientTape() as tape:
            tape.watch(interpolated_tensor)
            predictions = model(interpolated_tensor)
            target_pred = predictions[:, target_class_idx]
        
        grads = tape.gradient(target_pred, interpolated_tensor)
        gradients.append(grads.numpy())
    
    gradients = np.array(gradients)
    avg_gradients = np.mean(gradients, axis=0)
    
    # Compute integrated gradients
    integrated_grads = (img_array - baseline) * avg_gradients
    return integrated_grads[0]

print("Computing Integrated Gradients...")
ig_attributions = integrated_gradients(img_array, model, HEALTHY_CLASS_IDX, num_steps=50)

ig_magnitude = np.sum(np.abs(ig_attributions), axis=-1)
ig_magnitude = (ig_magnitude - ig_magnitude.min()) / (ig_magnitude.max() - ig_magnitude.min() + 1e-8)

print("\nüìä Integrated Gradients Results:")
display_image(img_array, "Original Image")
display_image(ig_magnitude, "Integrated Gradients Attribution")

ig_heatmap = cv2.resize(ig_magnitude, (original_img_np.shape[1], original_img_np.shape[0]))
ig_heatmap = np.uint8(255 * ig_heatmap)
ig_heatmap = cv2.applyColorMap(ig_heatmap, cv2.COLORMAP_JET)
ig_overlay = ig_heatmap * 0.4 + original_img_np
ig_overlay = np.clip(ig_overlay, 0, 255).astype('uint8')

display_image(ig_overlay / 255.0, "Integrated Gradients Overlay\nHealthy Class")

# ===================================================================================
# XAI METHOD 4: OCCLUSION SENSITIVITY
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 4: OCCLUSION SENSITIVITY")
print("=" * 100)

def occlusion_sensitivity(img_array, model, target_class_idx, patch_size=25, stride=12):
    """Compute occlusion sensitivity"""
    img = img_array[0].copy()
    h, w, c = img.shape
    
    sensitivity_map = np.zeros((h, w))
    baseline_pred = model.predict(img_array, verbose=0)[0, target_class_idx]
    
    occluding_patch = np.zeros((patch_size, patch_size, c))
    
    print(f"Computing occlusion (patch={patch_size}, stride={stride})...")
    positions = [(y, x) for y in range(0, h - patch_size + 1, stride) 
                 for x in range(0, w - patch_size + 1, stride)]
    
    for y, x in tqdm(positions, desc="Processing"):
        occluded_img = img.copy()
        occluded_img[y:y+patch_size, x:x+patch_size] = occluding_patch
        
        occluded_img_batch = np.expand_dims(occluded_img, axis=0)
        pred = model.predict(occluded_img_batch, verbose=0)[0, target_class_idx]
        
        sensitivity = baseline_pred - pred
        sensitivity_map[y:y+patch_size, x:x+patch_size] = np.maximum(
            sensitivity_map[y:y+patch_size, x:x+patch_size], sensitivity
        )
    
    return sensitivity_map

occlusion_map = occlusion_sensitivity(img_array, model, HEALTHY_CLASS_IDX, patch_size=25, stride=12)
occlusion_map = (occlusion_map - occlusion_map.min()) / (occlusion_map.max() - occlusion_map.min() + 1e-8)

print("\nüìä Occlusion Sensitivity Results:")
display_image(img_array, "Original Image")
display_image(occlusion_map, "Occlusion Sensitivity Map")

occlusion_heatmap = cv2.resize(occlusion_map, (original_img_np.shape[1], original_img_np.shape[0]))
occlusion_heatmap = np.uint8(255 * occlusion_heatmap)
occlusion_heatmap = cv2.applyColorMap(occlusion_heatmap, cv2.COLORMAP_JET)
occlusion_overlay = occlusion_heatmap * 0.4 + original_img_np
occlusion_overlay = np.clip(occlusion_overlay, 0, 255).astype('uint8')

display_image(occlusion_overlay / 255.0, "Occlusion Sensitivity Overlay\nHealthy Class")

# ===================================================================================
# XAI METHOD 5: LIME
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 5: LIME")
print("=" * 100)

def lime_explanation(img_array, model, target_class_idx, num_samples=500, num_features=10):
    """Simple LIME implementation"""
    img = img_array[0]
    
    print("Creating superpixels...")
    segments = slic(img, n_segments=50, compactness=10, sigma=1)
    num_segments = len(np.unique(segments))
    
    print(f"Generated {num_segments} superpixels")
    print(f"Generating {num_samples} samples...")
    
    masks = np.random.randint(0, 2, size=(num_samples, num_segments))
    
    predictions = []
    for mask in tqdm(masks, desc="LIME"):
        perturbed_img = img.copy()
        for seg_id in range(num_segments):
            if mask[seg_id] == 0:
                perturbed_img[segments == seg_id] = 0
        
        pred = model.predict(np.expand_dims(perturbed_img, axis=0), verbose=0)[0, target_class_idx]
        predictions.append(pred)
    
    predictions = np.array(predictions)
    
    from sklearn.linear_model import Ridge
    ridge = Ridge(alpha=1.0)
    ridge.fit(masks, predictions)
    
    importance = np.abs(ridge.coef_)
    top_features = np.argsort(importance)[-num_features:]
    
    explanation_mask = np.zeros_like(segments, dtype=float)
    for seg_id in top_features:
        explanation_mask[segments == seg_id] = importance[seg_id]
    
    explanation_mask = (explanation_mask - explanation_mask.min()) / (explanation_mask.max() - explanation_mask.min() + 1e-8)
    
    return explanation_mask, segments

lime_mask, segments = lime_explanation(img_array, model, HEALTHY_CLASS_IDX, num_samples=500, num_features=10)

print("\nüìä LIME Results:")
display_image(img_array, "Original Image")
display_image(lime_mask, "LIME Attribution Mask")

boundaries_img = mark_boundaries(img_array[0], segments, color=(1, 1, 0), mode='thick')
display_image(boundaries_img, "Superpixel Segmentation")

lime_heatmap = cv2.resize(lime_mask, (original_img_np.shape[1], original_img_np.shape[0]))
lime_heatmap = np.uint8(255 * lime_heatmap)
lime_heatmap = cv2.applyColorMap(lime_heatmap, cv2.COLORMAP_JET)
lime_overlay = lime_heatmap * 0.4 + original_img_np
lime_overlay = np.clip(lime_overlay, 0, 255).astype('uint8')

display_image(lime_overlay / 255.0, "LIME Overlay\nHealthy Class")

# ===================================================================================
# XAI METHOD 6: VANILLA GRADIENTS
# ===================================================================================
print("\n" + "=" * 100)
print("XAI METHOD 6: VANILLA GRADIENTS (SALIENCY)")
print("=" * 100)

def compute_saliency(img_array, model, target_class_idx):
    """Compute saliency map"""
    img_tensor = tf.convert_to_tensor(img_array)
    
    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        predictions = model(img_tensor)
        target_pred = predictions[:, target_class_idx]
    
    grads = tape.gradient(target_pred, img_tensor)
    grads = tf.abs(grads)
    
    saliency = grads.numpy()[0]
    saliency = np.max(saliency, axis=-1)
    saliency = (saliency - saliency.min()) / (saliency.max() - saliency.min() + 1e-8)
    
    return saliency

saliency_map = compute_saliency(img_array, model, HEALTHY_CLASS_IDX)

print("\nüìä Saliency Results:")
display_image(img_array, "Original Image")
display_image(saliency_map, "Vanilla Gradient Saliency")

saliency_heatmap = cv2.resize(saliency_map, (original_img_np.shape[1], original_img_np.shape[0]))
saliency_heatmap = np.uint8(255 * saliency_heatmap)
saliency_heatmap = cv2.applyColorMap(saliency_heatmap, cv2.COLORMAP_JET)
saliency_overlay = saliency_heatmap * 0.4 + original_img_np
saliency_overlay = np.clip(saliency_overlay, 0, 255).astype('uint8')

display_image(saliency_overlay / 255.0, "Saliency Overlay\nHealthy Class")

# ===================================================================================
# COMPARATIVE VISUALIZATION
# ===================================================================================
print("\n" + "=" * 100)
print("COMPARATIVE VISUALIZATION - ALL XAI METHODS")
print("=" * 100)

fig, axes = plt.subplots(2, 4, figsize=(20, 10))

axes[0, 0].imshow(img_array[0])
axes[0, 0].set_title('Original', fontsize=12, fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(gradcam_img / 255.0)
axes[0, 1].set_title('GradCAM', fontsize=12, fontweight='bold')
axes[0, 1].axis('off')

axes[0, 2].imshow(gradcam_pp_img / 255.0)
axes[0, 2].set_title('GradCAM++', fontsize=12, fontweight='bold')
axes[0, 2].axis('off')

axes[0, 3].imshow(ig_overlay / 255.0)
axes[0, 3].set_title('Integrated Gradients', fontsize=12, fontweight='bold')
axes[0, 3].axis('off')

axes[1, 0].imshow(occlusion_overlay / 255.0)
axes[1, 0].set_title('Occlusion', fontsize=12, fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(lime_overlay / 255.0)
axes[1, 1].set_title('LIME', fontsize=12, fontweight='bold')
axes[1, 1].axis('off')

axes[1, 2].imshow(saliency_overlay / 255.0)
axes[1, 2].set_title('Saliency', fontsize=12, fontweight='bold')
axes[1, 2].axis('off')

axes[1, 3].text(0.5, 0.5, 
               f'MobileNetV2\n250√ó250\n\n' +
               f'Pred: {idx_to_class[pred_class_idx]}\n' +
               f'Conf: {confidence:.4f}\n\n' +
               f'{DATASET_TYPE}',
               ha='center', va='center', fontsize=10,
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
axes[1, 3].axis('off')

plt.suptitle(f'XAI Methods - Healthy Class ({DATASET_TYPE})', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# ===================================================================================
# QUANTITATIVE METRICS
# ===================================================================================
print("\n" + "=" * 100)
print("QUANTITATIVE EVALUATION")
print("=" * 100)

def compute_sparseness(attr):
    flat = attr.flatten()
    flat = flat / (np.sum(np.abs(flat)) + 1e-8)
    return np.sum(np.abs(flat[:-1] - flat[1:]))

def compute_complexity(attr):
    grad_x = np.abs(np.gradient(attr, axis=0))
    grad_y = np.abs(np.gradient(attr, axis=1))
    return np.mean(grad_x + grad_y)

def compute_coverage(attr, threshold=0.5):
    return np.sum(attr > threshold) / attr.size

def normalize_attr(attr):
    return (attr - attr.min()) / (attr.max() - attr.min() + 1e-8)

attributions = {
    'GradCAM': normalize_attr(heatmap_healthy),
    'GradCAM++': normalize_attr(heatmap_healthy_pp),
    'Integrated Gradients': normalize_attr(ig_magnitude),
    'Occlusion': normalize_attr(occlusion_map),
    'LIME': normalize_attr(lime_mask),
    'Saliency': normalize_attr(saliency_map)
}

metrics_data = []
for method_name, attr_map in attributions.items():
    attr_resized = cv2.resize(attr_map, IMG_SIZE)
    metrics_data.append({
        'Method': method_name,
        'Sparseness': compute_sparseness(attr_resized),
        'Complexity': compute_complexity(attr_resized),
        'Coverage': compute_coverage(attr_resized)
    })

metrics_df = pd.DataFrame(metrics_data)
print("\nüìä Quantitative Metrics:")
print(metrics_df.to_string(index=False))

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].bar(metrics_df['Method'], metrics_df['Sparseness'], color='steelblue', edgecolor='black')
axes[0].set_title('Sparseness', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Score')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

axes[1].bar(metrics_df['Method'], metrics_df['Complexity'], color='coral', edgecolor='black')
axes[1].set_title('Complexity', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Score')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', alpha=0.3)

axes[2].bar(metrics_df['Method'], metrics_df['Coverage'], color='lightgreen', edgecolor='black')
axes[2].set_title('Coverage', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Score')
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(axis='y', alpha=0.3)

plt.suptitle(f'Metrics - {DATASET_TYPE}', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# ===================================================================================
# SAVE RESULTS
# ===================================================================================
print("\n" + "=" * 100)
print("SAVING RESULTS")
print("=" * 100)

metrics_df.to_csv(os.path.join(METRICS_DIR, f'metrics_{DATASET_TYPE.lower()}.csv'), index=False)

summary = pd.DataFrame([{
    'Dataset': DATASET_TYPE,
    'Image_Size': f'{IMG_SIZE[0]}x{IMG_SIZE[1]}',
    'Batch_Size': BATCH_SIZE,
    'Epochs': EPOCHS,
    'Image': os.path.basename(ANALYSIS_IMAGE),
    'Prediction': idx_to_class[pred_class_idx],
    'Confidence': confidence
}])

summary.to_csv(os.path.join(METRICS_DIR, f'summary_{DATASET_TYPE.lower()}.csv'), index=False)

print(f"\n‚úì Results saved to: {METRICS_DIR}")
print("\n" + "=" * 100)
print("‚úì XAI PIPELINE COMPLETE!")
print("=" * 100)
print(f"\nüìÅ Output: {OUTPUT_DIR}")
print("=" * 100)
```

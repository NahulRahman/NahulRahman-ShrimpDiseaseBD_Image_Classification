```python
#!/usr/bin/env python3
"""
===================================================================================
COMPLETE XAI PIPELINE - GAN AUGMENTED SHRIMP DATASET
===================================================================================
6 XAI Methods: GradCAM, GradCAM++, Integrated Gradients, Occlusion, LIME, Saliency
MobileNetV2 with 250x250 input, 30 epochs, batch size 64
Includes Confusion Matrix Evaluation
===================================================================================
"""

# ===================================================================================
# INSTALLATION
# ===================================================================================
"""
!pip install scikit-image opencv-python tqdm --quiet
"""

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import confusion_matrix, classification_report, auc as compute_auc
import cv2
from scipy.ndimage import gaussian_filter
import random
from skimage.segmentation import mark_boundaries, slic
from sklearn.linear_model import Ridge
from tqdm import tqdm
import time

np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

print("=" * 100)
print("COMPLETE XAI PIPELINE - GAN AUGMENTED DATASET")
print("=" * 100)

# ===================================================================================
# CONFIGURATION
# ===================================================================================

BATCH_SIZE = 64
EPOCHS = 30
IMG_SIZE = (250, 250)
NUM_CLASSES = 1  # Binary

# GAN DATASET
DATASET_TYPE = "GAN_AUGMENTED"
BASE_DIR = "/kaggle/input/elsevier-disease-shrimp-data/shrimp_healthy-styleGAN-ttvs"
AUGMENTATION_MARKER = "-stylegan-"

TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR = os.path.join(BASE_DIR, "validation")
TEST_DIR = os.path.join(BASE_DIR, "test")

TRAIN_HEALTHY = os.path.join(TRAIN_DIR, "Healthy")
VAL_HEALTHY = os.path.join(VAL_DIR, "Healthy")
TEST_HEALTHY = os.path.join(TEST_DIR, "Healthy")

OUTPUT_DIR = f"/home/claude/xai_complete_gan"
MODELS_DIR = os.path.join(OUTPUT_DIR, "models")
METRICS_DIR = os.path.join(OUTPUT_DIR, "metrics")
PLOTS_DIR = os.path.join(OUTPUT_DIR, "plots")

for directory in [OUTPUT_DIR, MODELS_DIR, METRICS_DIR, PLOTS_DIR]:
    os.makedirs(directory, exist_ok=True)

print(f"\nüìã CONFIGURATION:")
print(f"   Dataset: {DATASET_TYPE}")
print(f"   Base: {BASE_DIR}")
print(f"   Image Size: {IMG_SIZE}")

# ===================================================================================
# UTILITIES
# ===================================================================================

def get_healthy_images(directory, augmentation_marker=None, n_samples=None):
    if not os.path.exists(directory):
        return []
    all_files = [f for f in os.listdir(directory) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    if augmentation_marker:
        all_files = [f for f in all_files if augmentation_marker in f]
    if n_samples:
        all_files = random.sample(all_files, min(n_samples, len(all_files)))
    return [os.path.join(directory, f) for f in all_files]

def load_and_preprocess_image(img_path, target_size=IMG_SIZE):
    img = load_img(img_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0
    return img_array, img

def display_image(img_array, title="", figsize=(6, 6)):
    plt.figure(figsize=figsize)
    if len(img_array.shape) == 4:
        img_array = img_array[0]
    plt.imshow(img_array if img_array.shape[-1] == 3 else img_array, cmap=None if img_array.shape[-1] == 3 else 'gray')
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# ===================================================================================
# LOAD/TRAIN MODEL
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 1: MOBILENETV2 BINARY CLASSIFICATION")
print("=" * 100)

model_path = os.path.join(MODELS_DIR, 'mobilenetv2_gan_250x250.h5')

if os.path.exists(model_path):
    print(f"Loading: {model_path}")
    model = load_model(model_path)
    print("‚úì Loaded")
else:
    print("Training...")
    
    train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=20, width_shift_range=0.2,
                                       height_shift_range=0.2, horizontal_flip=True, fill_mode='nearest')
    val_datagen = ImageDataGenerator(rescale=1./255)
    
    train_generator = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE, 
                                                        batch_size=BATCH_SIZE, class_mode='binary')
    val_generator = val_datagen.flow_from_directory(VAL_DIR, target_size=IMG_SIZE,
                                                    batch_size=BATCH_SIZE, class_mode='binary')
    
    print(f"Train: {train_generator.samples}, Val: {val_generator.samples}")
    
    base_model = MobileNetV2(input_shape=IMG_SIZE + (3,), include_top=False, weights='imagenet')
    for layer in base_model.layers:
        layer.trainable = False
    
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    predictions = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
    
    print(f"Parameters: {model.count_params():,}")
    
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7)
    ]
    
    print("üöÄ Training...")
    model.fit(train_generator, validation_data=val_generator, epochs=EPOCHS, callbacks=callbacks, verbose=1)
    model.save(model_path)
    print(f"‚úì Saved")

# ===================================================================================
# CONFUSION MATRIX
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 2: CONFUSION MATRIX")
print("=" * 100)

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE, batch_size=1, 
                                                  class_mode='binary', shuffle=False)

print(f"Evaluating {test_generator.samples} samples...")

y_true = test_generator.classes
y_pred_probs = model.predict(test_generator, verbose=1)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

class_names = list(test_generator.class_indices.keys())
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix - GAN Dataset', fontsize=14, fontweight='bold')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'confusion_matrix.png'), dpi=300, bbox_inches='tight')
plt.show()

print("\nüìä Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

accuracy = np.mean(y_true == y_pred)
print(f"‚úì Accuracy: {accuracy:.4f}")

# ===================================================================================
# SELECT IMAGES
# ===================================================================================
print("\n" + "=" * 100)
print("STEP 3: SELECT IMAGES")
print("=" * 100)

all_healthy = get_healthy_images(VAL_HEALTHY, AUGMENTATION_MARKER)
if not all_healthy:
    all_healthy = get_healthy_images(TEST_HEALTHY, AUGMENTATION_MARKER)
if not all_healthy:
    all_healthy = get_healthy_images(TRAIN_HEALTHY, AUGMENTATION_MARKER)

ANALYSIS_IMAGE = random.choice(all_healthy) if all_healthy else None

if ANALYSIS_IMAGE is None:
    print("‚ö† No images. Exiting.")
    exit()

print(f"üì∏ Selected: {os.path.basename(ANALYSIS_IMAGE)}")

# ===================================================================================
# XAI METHOD 1: GRADCAM
# ===================================================================================
print("\n" + "=" * 100)
print("XAI 1: GRAD-CAM")
print("=" * 100)

def get_last_conv_layer(model):
    for layer in reversed(model.layers):
        if 'conv' in layer.name.lower():
            return layer.name
    return None

def make_gradcam_heatmap(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(last_conv_layer_name).output, model.output])
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        class_channel = predictions[:, 0]  # Binary: single output
    
    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def apply_gradcam(img_array, heatmap, original_img):
    if isinstance(original_img, np.ndarray):
        if original_img.max() <= 1.0:
            original_img = original_img * 255
        original_img = original_img.astype('uint8')
    
    target_h, target_w = original_img.shape[:2]
    heatmap_resized = cv2.resize(heatmap, (target_w, target_h))
    heatmap_resized = np.uint8(255 * heatmap_resized)
    heatmap_colored = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)
    
    superimposed = heatmap_colored * 0.4 + original_img
    superimposed = np.clip(superimposed, 0, 255).astype('uint8')
    return superimposed

last_conv = get_last_conv_layer(model)
print(f"Last conv: {last_conv}")

img_array, original_img = load_and_preprocess_image(ANALYSIS_IMAGE)
original_img_np = np.array(original_img)

pred = model.predict(img_array, verbose=0)[0][0]
print(f"Prediction: {pred:.4f}")

heatmap = make_gradcam_heatmap(img_array, model, last_conv)
gradcam_img = apply_gradcam(img_array[0], heatmap, original_img_np)

display_image(img_array, "Original")
display_image(gradcam_img / 255.0, f"GradCAM\nProb: {pred:.4f}")

# ===================================================================================
# XAI 2-6: Similar implementations (abbreviated for space)
# ===================================================================================
# I'll show the structure but keep it concise

print("\n" + "=" * 100)
print("XAI 2: GRAD-CAM++")
print("=" * 100)

def make_gradcam_plusplus_heatmap(img_array, model, last_conv_layer_name):
    grad_model = tf.keras.models.Model([model.inputs], [model.get_layer(last_conv_layer_name).output, model.output])
    
    with tf.GradientTape() as tape1:
        with tf.GradientTape() as tape2:
            with tf.GradientTape() as tape3:
                conv_outputs, predictions = grad_model(img_array)
                class_channel = predictions[:, 0]
            grads = tape3.gradient(class_channel, conv_outputs)
        grads_2 = tape2.gradient(grads, conv_outputs)
    grads_3 = tape1.gradient(grads_2, conv_outputs)
    
    grads = grads[0]
    grads_2 = grads_2[0]
    grads_3 = grads_3[0]
    
    global_sum = tf.reduce_sum(conv_outputs[0], axis=(0, 1), keepdims=True)
    alpha_denom = grads_2 * 2.0 + grads_3 * global_sum
    alpha_denom = tf.where(alpha_denom != 0.0, alpha_denom, tf.ones_like(alpha_denom))
    alphas = grads_2 / alpha_denom
    
    weights = tf.reduce_sum(alphas * tf.nn.relu(grads), axis=(0, 1))
    
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ weights[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.nn.relu(heatmap)
    heatmap = heatmap / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

heatmap_pp = make_gradcam_plusplus_heatmap(img_array, model, last_conv)
gradcam_pp_img = apply_gradcam(img_array[0], heatmap_pp, original_img_np)

display_image(gradcam_pp_img / 255.0, f"GradCAM++\nProb: {pred:.4f}")

# Integrated Gradients
print("\n" + "=" * 100)
print("XAI 3: INTEGRATED GRADIENTS")
print("=" * 100)

def integrated_gradients(img_array, model, baseline=None, num_steps=50):
    if baseline is None:
        baseline = np.zeros_like(img_array)
    
    alphas = np.linspace(0, 1, num_steps)
    gradients = []
    
    for alpha in tqdm(alphas, desc="IG"):
        interpolated = baseline + alpha * (img_array - baseline)
        interpolated_tensor = tf.convert_to_tensor(interpolated, dtype=tf.float32)
        
        with tf.GradientTape() as tape:
            tape.watch(interpolated_tensor)
            predictions = model(interpolated_tensor)
            target_pred = predictions[:, 0]
        
        grads = tape.gradient(target_pred, interpolated_tensor)
        gradients.append(grads.numpy())
    
    gradients = np.array(gradients)
    avg_gradients = np.mean(gradients, axis=0)
    
    integrated_grads = (img_array - baseline) * avg_gradients
    return integrated_grads[0]

ig_attributions = integrated_gradients(img_array, model, num_steps=50)
ig_magnitude = np.sum(np.abs(ig_attributions), axis=-1)
ig_magnitude = (ig_magnitude - ig_magnitude.min()) / (ig_magnitude.max() - ig_magnitude.min() + 1e-8)

display_image(ig_magnitude, "Integrated Gradients")

ig_heatmap = cv2.resize(ig_magnitude, (original_img_np.shape[1], original_img_np.shape[0]))
ig_heatmap = np.uint8(255 * ig_heatmap)
ig_heatmap = cv2.applyColorMap(ig_heatmap, cv2.COLORMAP_JET)
ig_overlay = ig_heatmap * 0.4 + original_img_np
ig_overlay = np.clip(ig_overlay, 0, 255).astype('uint8')

display_image(ig_overlay / 255.0, "IG Overlay")

# Occlusion
print("\n" + "=" * 100)
print("XAI 4: OCCLUSION SENSITIVITY")
print("=" * 100)

def occlusion_sensitivity(img_array, model, patch_size=25, stride=12):
    img = img_array[0].copy()
    h, w, c = img.shape
    
    sensitivity_map = np.zeros((h, w))
    baseline_pred = model.predict(img_array, verbose=0)[0, 0]
    
    occluding_patch = np.zeros((patch_size, patch_size, c))
    
    positions = [(y, x) for y in range(0, h - patch_size + 1, stride) 
                 for x in range(0, w - patch_size + 1, stride)]
    
    for y, x in tqdm(positions, desc="Occlusion"):
        occluded_img = img.copy()
        occluded_img[y:y+patch_size, x:x+patch_size] = occluding_patch
        
        occluded_img_batch = np.expand_dims(occluded_img, axis=0)
        pred = model.predict(occluded_img_batch, verbose=0)[0, 0]
        
        sensitivity = baseline_pred - pred
        sensitivity_map[y:y+patch_size, x:x+patch_size] = np.maximum(
            sensitivity_map[y:y+patch_size, x:x+patch_size], sensitivity
        )
    
    return sensitivity_map

occlusion_map = occlusion_sensitivity(img_array, model, patch_size=25, stride=12)
occlusion_map = (occlusion_map - occlusion_map.min()) / (occlusion_map.max() - occlusion_map.min() + 1e-8)

display_image(occlusion_map, "Occlusion Sensitivity")

occlusion_heatmap = cv2.resize(occlusion_map, (original_img_np.shape[1], original_img_np.shape[0]))
occlusion_heatmap = np.uint8(255 * occlusion_heatmap)
occlusion_heatmap = cv2.applyColorMap(occlusion_heatmap, cv2.COLORMAP_JET)
occlusion_overlay = occlusion_heatmap * 0.4 + original_img_np
occlusion_overlay = np.clip(occlusion_overlay, 0, 255).astype('uint8')

display_image(occlusion_overlay / 255.0, "Occlusion Overlay")

# LIME
print("\n" + "=" * 100)
print("XAI 5: LIME")
print("=" * 100)

def lime_explanation(img_array, model, num_samples=500, num_features=10):
    img = img_array[0]
    
    segments = slic(img, n_segments=50, compactness=10, sigma=1)
    num_segments = len(np.unique(segments))
    
    print(f"Superpixels: {num_segments}")
    
    masks = np.random.randint(0, 2, size=(num_samples, num_segments))
    
    predictions = []
    for mask in tqdm(masks, desc="LIME"):
        perturbed_img = img.copy()
        for seg_id in range(num_segments):
            if mask[seg_id] == 0:
                perturbed_img[segments == seg_id] = 0
        
        pred = model.predict(np.expand_dims(perturbed_img, axis=0), verbose=0)[0, 0]
        predictions.append(pred)
    
    predictions = np.array(predictions)
    
    ridge = Ridge(alpha=1.0)
    ridge.fit(masks, predictions)
    
    importance = np.abs(ridge.coef_)
    top_features = np.argsort(importance)[-num_features:]
    
    explanation_mask = np.zeros_like(segments, dtype=float)
    for seg_id in top_features:
        explanation_mask[segments == seg_id] = importance[seg_id]
    
    explanation_mask = (explanation_mask - explanation_mask.min()) / (explanation_mask.max() - explanation_mask.min() + 1e-8)
    
    return explanation_mask, segments

lime_mask, segments = lime_explanation(img_array, model, num_samples=500, num_features=10)

display_image(lime_mask, "LIME Attribution")

boundaries_img = mark_boundaries(img_array[0], segments, color=(1, 1, 0), mode='thick')
display_image(boundaries_img, "Superpixels")

lime_heatmap = cv2.resize(lime_mask, (original_img_np.shape[1], original_img_np.shape[0]))
lime_heatmap = np.uint8(255 * lime_heatmap)
lime_heatmap = cv2.applyColorMap(lime_heatmap, cv2.COLORMAP_JET)
lime_overlay = lime_heatmap * 0.4 + original_img_np
lime_overlay = np.clip(lime_overlay, 0, 255).astype('uint8')

display_image(lime_overlay / 255.0, "LIME Overlay")

# Saliency
print("\n" + "=" * 100)
print("XAI 6: VANILLA GRADIENTS")
print("=" * 100)

def compute_saliency(img_array, model):
    img_tensor = tf.convert_to_tensor(img_array)
    
    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        predictions = model(img_tensor)
        target_pred = predictions[:, 0]
    
    grads = tape.gradient(target_pred, img_tensor)
    grads = tf.abs(grads)
    
    saliency = grads.numpy()[0]
    saliency = np.max(saliency, axis=-1)
    saliency = (saliency - saliency.min()) / (saliency.max() - saliency.min() + 1e-8)
    
    return saliency

saliency_map = compute_saliency(img_array, model)

display_image(saliency_map, "Vanilla Gradients")

saliency_heatmap = cv2.resize(saliency_map, (original_img_np.shape[1], original_img_np.shape[0]))
saliency_heatmap = np.uint8(255 * saliency_heatmap)
saliency_heatmap = cv2.applyColorMap(saliency_heatmap, cv2.COLORMAP_JET)
saliency_overlay = saliency_heatmap * 0.4 + original_img_np
saliency_overlay = np.clip(saliency_overlay, 0, 255).astype('uint8')

display_image(saliency_overlay / 255.0, "Saliency Overlay")

# ===================================================================================
# METRICS
# ===================================================================================
print("\n" + "=" * 100)
print("QUANTITATIVE METRICS")
print("=" * 100)

def compute_sparseness(attr):
    flat = attr.flatten()
    flat = flat / (np.sum(np.abs(flat)) + 1e-8)
    return np.sum(np.abs(flat[:-1] - flat[1:]))

def compute_complexity(attr):
    grad_x = np.abs(np.gradient(attr, axis=0))
    grad_y = np.abs(np.gradient(attr, axis=1))
    return np.mean(grad_x + grad_y)

def compute_coverage(attr, threshold=0.5):
    return np.sum(attr > threshold) / attr.size

def normalize_attr(attr):
    return (attr - attr.min()) / (attr.max() - attr.min() + 1e-8)

attributions = {
    'GradCAM': normalize_attr(heatmap),
    'GradCAM++': normalize_attr(heatmap_pp),
    'Integrated Gradients': normalize_attr(ig_magnitude),
    'Occlusion': normalize_attr(occlusion_map),
    'LIME': normalize_attr(lime_mask),
    'Saliency': normalize_attr(saliency_map)
}

metrics_data = []
for method_name, attr_map in attributions.items():
    attr_resized = cv2.resize(attr_map, IMG_SIZE)
    
    metrics_data.append({
        'Method': method_name,
        'Sparseness': compute_sparseness(attr_resized),
        'Complexity': compute_complexity(attr_resized),
        'Coverage': compute_coverage(attr_resized)
    })

metrics_df = pd.DataFrame(metrics_data)
print("\nüìä Metrics:")
print(metrics_df.to_string(index=False))

metrics_df.to_csv(os.path.join(METRICS_DIR, 'xai_metrics.csv'), index=False)

# Visualize
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].bar(metrics_df['Method'], metrics_df['Sparseness'], edgecolor='black')
axes[0].set_title('Sparseness', fontsize=13, fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

axes[1].bar(metrics_df['Method'], metrics_df['Complexity'], edgecolor='black', color='coral')
axes[1].set_title('Complexity', fontsize=13, fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', alpha=0.3)

axes[2].bar(metrics_df['Method'], metrics_df['Coverage'], edgecolor='black', color='lightgreen')
axes[2].set_title('Coverage', fontsize=13, fontweight='bold')
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(PLOTS_DIR, 'metrics_comparison.png'), dpi=300, bbox_inches='tight')
plt.show()

summary = pd.DataFrame([{
    'Dataset': DATASET_TYPE,
    'Image': os.path.basename(ANALYSIS_IMAGE),
    'Prediction_Prob': pred,
    'Test_Accuracy': accuracy
}])

summary.to_csv(os.path.join(METRICS_DIR, 'summary.csv'), index=False)

print("\n" + "=" * 100)
print("‚úì COMPLETE!")
print("=" * 100)
print(f"üìÅ Output: {OUTPUT_DIR}")
print(f"üì∏ Image: {os.path.basename(ANALYSIS_IMAGE)}")
print(f"üéØ Prob: {pred:.4f}")
print(f"üéØ Test Accuracy: {accuracy:.4f}")
print("=" * 100)
```
